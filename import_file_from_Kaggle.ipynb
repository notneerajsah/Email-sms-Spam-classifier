```python
import kagglehub
import shutil
import os
import pandas as pd

# 1. Download the latest version of the dataset
source_path = kagglehub.dataset_download("uciml/sms-spam-collection-dataset")

# 2. Define your working directory (replace with your path)
working_dir = r"C:\Users\iamne\Desktop\New folder\sms-classifier"

# 3. Define destination path for dataset
dest_path = os.path.join(working_dir, "sms-spam-collection-dataset")

# 4. Copy dataset files into your working directory
if not os.path.exists(dest_path):
    shutil.copytree(source_path, dest_path)
    print(f"✅ Dataset copied to: {dest_path}")
else:
    print(f"ℹ️ Dataset already exists at: {dest_path}")

# 5. Load the dataset (adjust file name if needed)
csv_file = os.path.join(dest_path, "spam.csv")

if os.path.exists(csv_file):
    df = pd.read_csv(csv_file, encoding='latin-1')
    print("✅ Dataset loaded successfully!")
    print(df.head())
else:
    print("⚠️ Could not find 'spam.csv' in the dataset folder.")
```
